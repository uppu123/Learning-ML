### There are 2 techniques to handle imbalaced datasets:
### 1. Up - Sampling the minority class
### 2. Down - sampling the majority class

```python
import numpy as np
import pandas as pd

np.random.seed(0)

n_samples = 1000
class_0_ratio = 0.9

nclass_0 = int(n_samples * class_0_ratio)
nclass_1 = n_samples - nclass_0
```

```python
### Creating Dataframe with imbalaced dataset

class_0 = pd.DataFrame({
    'feature 1' : np.random.normal(loc = 0, scale = 1, size = nclass_0),
    'feature 2' : np.random.normal(loc = 0, scale = 1, size = nclass_0),
    'target' : [0] * nclass_0
})

class_1 = pd.DataFrame({
    'feature 1' : np.random.normal(loc = 2, scale = 1, size = nclass_1),
    'feature 2' : np.random.normal(loc = 2, scale = 1, size = nclass_1),
    'target' : [1] * nclass_1
})

df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
### Up - sampling the minority class

from sklearn.utils import resample
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]

df_minority.shape
df_majority.shape
```

```python
df_minority_upsampled = resample(df_minority, replace = True, n_samples = len(df_majority), random_state = 42)

df_upsampled = pd.concat([df_majority, df_minority_upsampled]).reset_index(drop = True) 
```

```python
df_upsampled.shape  ## 900 - 1's, 900 - 0's
```

```python
### Down - Sampling the majority class

df_majority_downsampled = resample(df_majority, replace = False, n_samples = len(df_minority), random_state = 42)

df_downsampled = pd.concat([df_majority_downsampled, df_minority]).reset_index(drop = True)
```

```python
df_downsampled.shape    ## 200 - 1's, 200 - 0's
```
