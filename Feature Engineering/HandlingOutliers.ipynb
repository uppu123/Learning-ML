HANDLING OUTLIERS USING PYTHON IQR METHOD

### with using box plot and 5 numbers summary

```python
import numpy as np

lst_marks = [10,20,30,43,35,37,364,24,423,76,87,34,76,90]
```

```python
min, Q1, median, Q3, max = np.quantile(lst_marks, [0, 0.25, 0.5, 0.75, 1])

IQR = Q3 - Q1

lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR
```
