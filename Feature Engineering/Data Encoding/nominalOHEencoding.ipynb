### Encoding : categorical data -------------> Numerical data

Nominal / One Hot Encoding(OHE)

in this each category is represented as a Binary Vector Where each bit corresponds to a unique Category

```python
import pandas as pd
df = pd.DataFrame({
    'color' : ['red', 'blue', 'green', 'red', 'blue']
})
```

```python
### Creating and instance of ohe
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
# perform fit and transform

encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out(['color']))
```

```python
encoded_df
```
